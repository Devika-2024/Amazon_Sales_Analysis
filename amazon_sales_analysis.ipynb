```python
# Chart 1

# Monthly Sales with Bar chart

monthly_sales = df.groupby('Month_Name')['Amount'].sum()

month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

monthly_sales = monthly_sales.reindex(month_order).dropna()

monthly_sales.plot(kind='bar', figsize=(10,5))
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.savefig(f"{save_path}\\chart1_monthly_sales.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Monthly sales with  line chart

monthly_sales = df.groupby(['Month','Month_Name'])['Amount'].sum().reset_index()
monthly_sales = monthly_sales.sort_values('Month')

plt.figure(figsize=(10,5))
plt.plot(monthly_sales['Month_Name'], monthly_sales['Amount'], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.savefig(f"{save_path}\\chart1L_monthly_sales.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Chart 2

# Top Categories

top_cat = df.groupby('Category')['Amount'].sum().sort_values(ascending=False).head(10)

top_cat.plot(kind='bar', figsize=(10,5))
plt.title("Top Categories by Revenue")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.savefig(f"{save_path}\\chart2_top_categories.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Chart 3

#Top 10 states

top_states = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(10)

top_states.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 States by Revenue")
plt.xlabel("ship-state")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.savefig(f"{save_path}\\chart3_top_states.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Chart 4

# Order status checking

df['Status'].value_counts().head(10).plot(kind='bar')
plt.title("Order Status Distribution")
plt.savefig(f"{save_path}\\chart4_order_status.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df['Status'].value_counts()
```

```python
def simplify_status(x):
    if 'Cancelled' in x:
        return 'Cancelled'
    elif 'Pending' in x:
        return 'Pending'
    elif 'Returned' in x or 'Returning' in x or 'Rejected' in x:
        return 'Returned'
    elif 'Delivered' in x:
        return 'Delivered'
    elif 'Shipped' in x:
        return 'Shipped'
    else:
        return 'Other'

df['Status_Group'] = df['Status'].apply(simplify_status)

df['Status_Group'].value_counts().plot(kind='bar', figsize=(8,5))

plt.title("Order Status Summary")
plt.xlabel("Status")
plt.ylabel("Orders")
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig("chart4_order_status.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Chart 5

# Fulfillment Method

df['Fulfilment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Fulfillment Method Share")
plt.ylabel("")
plt.savefig(f"{save_path}\\chart5_fulfillment.png", dpi=300, bbox_inches="tight")
plt.show()
```
